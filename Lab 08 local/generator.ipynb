```python
import numpy as np
import pandas as pd

np.random.seed(42)

n = 100

# Hidden "academic ability" factor
academic_ability = np.random.normal(75, 10, n)

# Correlated academic subjects
math = academic_ability + np.random.normal(0, 5, n)
science = academic_ability + np.random.normal(0, 5, n)
english = academic_ability + np.random.normal(0, 5, n)

# Unrelated music ability
music = np.random.normal(70, 10, n)

df = pd.DataFrame({
    'Math': math,
    'Science': science,
    'English': english,
    'Music': music
})

df.head()
df.to_csv('grades.csv', index=False)
```
